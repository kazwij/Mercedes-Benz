## Variance checking and dropping

* get the variance from train_df
* select only the zero columns --> get column index
* drop these columns from train_df and test_df both. 

```python
#get the variance zero columns from train_df.
variance = train_df.var()
columns_to_drop = variance[variance==0].index
print(f'total number of columns dropped: {len(columns_to_drop)}')
train_df = train_df.drop(columns_to_drop,axis=1)
```

```python
## dropping same columns from test_df
test_df = test_df.drop(columns_to_drop,axis=1)
```

## Apply label encoder.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def ecodingdf(df):
    #select only object type columns and encode
    encode_df = df.select_dtypes(include=['object']).apply(LabelEncoder().fit_transform)
    #drop the original columns with object type and join encoded colums to df
    df_new  = df.drop(df.select_dtypes(['object']).columns,axis=1).join(encode_df)
    return df_new
```

```python
train_df  = ecodingdf(train_df)
```

```python
train_df.head()
```

```python
test_df = ecodingdf(test_df)
```

```python
test_df.head()
```

```python
train_features = train_df.drop('y',axis=1 )
```

```python
train_target = train_df['y']
```

## Perform dimensionality reduction.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.95) 
```

```python
type(train_features)
```

```python
reduced_data=pca.fit_transform(train_features)
```

```python
reduced_data.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
X_train = reduced_data
y_train = train_target
X_test =pca.fit_transform(test_df) # apply pca to test dataset also. 
```

## Predict  test_df values using XGBoost.

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(reduced_data,train_target,train_size=0.8,random_state=50)
```

```python
X_test
```

```python
model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
# Predicting y values based on test_df
predictions = model.predict(X_test)
y_pred = pd.DataFrame(predictions)
```

### Model tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {
    
    'n_estimators':[100,200,300], 
    'learning_rate':[0.05,0.1,0.2], 
    'max_depth':[3,5,7],
      
    
}
```

```python
grid_search = GridSearchCV(model, parameter, cv=3, scoring='neg_mean_squared_error') 
grid_search.fit(X_train, y_train)
```

```python
# Best parameters and best score
print(grid_search.best_params_)
print(grid_search.best_score_)
```
